```python
gdf = gpd.read_file("GeoDS4Bolivia.geojson")
print(gdf.info())
# """
# 70	sdg16_6_pbec	Programmed budget execution capacity, 2017 (%)
# 68	sdg15_5_blr	    Biodiversity loss rate due to deforestation deforestation deforestation, average 2016-2018"
# 66	sdg13_2_dra	    Deforestation rate, average 2016-2018 (% of fo...
# 65	sdg13_2_tco2e	Total CO2 emissions per capita, 2016 (tCO2/per...
# 59    sdg10_2_iec	    Inequality in electricity consumption, 2016
# 56	sdg9_c_mnc	    Fixed and mobile network coverage, 2012 (% of ...
# 53	sdg9_1_routes	Number of railways/primary roads entering/leaving the municipality, 2019"
# 51	sdg8_10_dbb	    Density of bank branches, 2018 (per 100,000 in...
# 45	sdg7_3_co2epc	CO2 emissions per capita by energy, 2016. (tCO...
# 43	sdg7_1_rec	    Residential electricity consumption per capita...
# 42	sdg7_1_ec	    Electricity coverage, 2012 (% of population) population)"
# """
```

```python
"""
42 sdg7_1_ec: Electricity coverage, 2012 (% of population)
43 sdg7_1_rec: Residential electricity consumption per capita, 2016 (kWh/person/year)
44 sdg7_1_cce: Clean cooking energy, 2012 (% of households)
45 sdg7_3_co2epc: CO2 emissions per capita by energy, 2016. (tCO2/person/year)
"""
df = gdf[["sdg10_2_iec", "sdg7_3_co2epc", "sdg7_1_rec", "sdg7_1_ec", "sdg7_1_cce", "dep", "geometry"]]
print(df.info())
# Column rename for easy understanding 
df.rename(columns={'sdg10_2_iec':'elec_consum_inequal', 'sdg7_3_co2epc': 'CO2_percapita', 'sdg7_1_rec':'elect_consumption', 'sdg7_1_ec':'elec_covarage', 'sdg7_1_cce':'clean_cookingenergy'}, inplace=True)
print(df.info())
print()
print(df.isnull().sum())
print()
print(df.describe())
```

```python
# Disterbution of the regons in Bolivia 
fig, axs = plt.subplots(ncols=1, figsize=(15, 5))
sns.histplot(data=df, x="dep", kde=True)
plt.show()
```

```python
# CO2 emission for each city
# Group the data by city and calculate the average CO2 emissions per capita for each city
co2_by_city = df.groupby('dep')['CO2_percapita'].mean()

# Sort the data in descending order of CO2 emissions per capita
co2_by_city_sorted = co2_by_city.sort_values(ascending=False)

# Display all cities with the highest CO2 emissions per capita 
print(co2_by_city_sorted.head(9)) # '9' as there are only 9 cities in bolivia

fig, axs = plt.subplots(ncols=1, figsize=(15, 5))
sns.barplot(x=co2_by_city_sorted.index, y=co2_by_city_sorted.values)
plt.xlabel('City')
# plt.xticks(rotation=45)
plt.ylabel('CO2 emissions per capita')
plt.title('Average CO2 Emissions per Capita by City')
plt.show()
```

```python
# Map of elec_consum_inequal

df.plot(column='elec_consum_inequal', cmap='OrRd', k=5, legend=True, figsize=(15, 10))
# gdf.plot(column='population', cmap='OrRd', legend=True, legend_kwds={'label': 'Population'}, scheme='quantiles', edgecolor='black', linewidth=0.5, figsize=(10, 10))
# Show the plot
plt.show()
```

```python
""" 
 1   CO2_percapita        339 non-null    float64 
 2   elect_consumption    339 non-null    int64   
 3   elec_covarage        339 non-null    float64 
 4   clean_cookingenergy  339 non-null    int64  
"""
df.plot(column='CO2_percapita', cmap='OrRd', k=5, legend=True, figsize=(15, 10))
df.plot(column='elect_consumption', cmap='OrRd', k=5, legend=True, figsize=(15, 10))
df.plot(column='elec_covarage', cmap='OrRd', k=5, legend=True, figsize=(15, 10))
df.plot(column='clean_cookingenergy', cmap='OrRd', k=5, legend=True, figsize=(15, 10))
```

```python
# Transform 'elec_consum_inequal' from object to numeric datatype
df['elec_consum_inequal'] = pd.to_numeric(df['elec_consum_inequal'], errors='coerce', downcast ='signed')
df = df.dropna()
print(df.isnull().sum())
print(df.info())
```

```python
# Electrical consumption sorted by city
# Group the data by city and calculate the average CO2 emissions per capita for each city
elec_consum_by_city = df.groupby('dep')['elect_consumption'].mean()

# Sort the data in descending order of CO2 emissions per capita
elec_consum_by_city_sorted = elec_consum_by_city.sort_values(ascending=False)

# Display all cities with the highest CO2 emissions per capita 
print(elec_consum_by_city_sorted.head(9)) # '9' as there are only 9 cities in bolivia

fig, axs = plt.subplots(ncols=1, figsize=(15, 5))
sns.barplot(x=elec_consum_by_city_sorted.index, y=elec_consum_by_city_sorted.values)
plt.xlabel('City')
# plt.xticks(rotation=45)
plt.ylabel('Electrical consumption')
plt.title('Average Electrical consumption by City')
plt.show()
```

```python
# Electrical covarage in each city
# Group the data by city and calculate the average CO2 emissions per capita for each city
elec_covarage_by_city = df.groupby('dep')['elec_covarage'].mean()

# Sort the data in descending order of CO2 emissions per capita
elec_covarage_by_city_sorted = elec_covarage_by_city.sort_values(ascending=False)

# Display all cities with the highest CO2 emissions per capita 
print(elec_covarage_by_city_sorted.head(9)) # '9' as there are only 9 cities in bolivia

fig, axs = plt.subplots(ncols=1, figsize=(15, 5))
sns.barplot(x=elec_covarage_by_city_sorted.index, y=elec_covarage_by_city_sorted.values)
plt.xlabel('City')
# plt.xticks(rotation=45)
plt.ylabel('Electrical covarage')
plt.title('Average Electrical covarage by City')
plt.show()
```

```python
# Electrical covarage in each city
# Group the data by city and calculate the average CO2 emissions per capita for each city
clean_cookingenergy_by_city = df.groupby('dep')['clean_cookingenergy'].mean()

# Sort the data in descending order of CO2 emissions per capita
clean_cookingenergy_by_city_sorted = clean_cookingenergy_by_city.sort_values(ascending=False)

# Display all cities with the highest CO2 emissions per capita 
print(clean_cookingenergy_by_city_sorted.head(9)) # '9' as there are only 9 cities in bolivia

fig, axs = plt.subplots(ncols=1, figsize=(15, 5))
sns.barplot(x=clean_cookingenergy_by_city_sorted.index, y=clean_cookingenergy_by_city_sorted.values)
plt.xlabel('City')
# plt.xticks(rotation=45)
plt.ylabel('Clean Cookingenergy')
plt.title('Average Clean Cookingenergy by City')
plt.show()
```
